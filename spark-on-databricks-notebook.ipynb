```
titanic_df = spark.read.csv('dbfs:/databricks-datasets/Rdatasets/data-001/csv/datasets/Titanic.csv', header=True, inferSchema=True)
```

```
display(titanic_df)
```

```
survived_titanic_df = titanic_df.filter(titanic_df.Survived != 'No')
```

```
display(survived_titanic_df)
```

```
survived_titanic_df.write.format('csv').mode('overwrite').option('header', 'True').save('dbfs:/titanic/survived/')
```
